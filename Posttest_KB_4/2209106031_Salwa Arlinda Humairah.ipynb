### 2. Normalisasi dan Standarisasi Data

```python
from sklearn.preprocessing import MinMaxScaler # untuk normalisasi MinMax data
from sklearn.preprocessing import StandardScaler # untuk standarisasi data
import numpy as np 
```

```python
new_df = df.copy() # Copy dataset
new_df
```

```python
mms = MinMaxScaler()
ss = StandardScaler()
```

```python
new_df.isna().sum() 

normalisasiData = mms.fit_transform(new_df[['Customer_rating']])
standarisasiData = ss.fit_transform(new_df[['Weight_in_gms', 'Discount_offered']])
```

#### Normalisasi data pada atribut 'Customer Rating' dan Nilai standar deviasi bernilai 1 karena pada atribut 'Customer Rating' dan 'Cost of the Product' tidak ada nilai null di dalam datanya.

### 3. Data Cleaning pada nilai null

#### Membuat nilai not null jadi null, karena di dataset ini tidak ada nilai null

```python
nilaiNotnull = new_df.iloc[8275:, :1] = np.nan # Membuat nilai null pada kolom 'ID'
nilaiNotnull = new_df.iloc[10000:, 4] = np.nan # Membuat nilai null pada kolom 'Customer Rating'
nilaiNotnull = new_df.iloc[1500:, 7] = np.nan # Membuat nilai null pada kolom 'Prior Purchases'
new_df.isna().sum()
```

#### Data Cleaning pada atribut 'ID'

##### Menggunakan median karena atribut ID bertipe data int atau bil. bulat

```python
new_df['ID'].fillna(new_df['ID'].median(), inplace = True)
new_df.isna().sum()
```

#### Data Cleaning pada atribut 'Customer Rating'

##### Menggunakan mean karena atribut Customer Rating bertipe data float atau desimal

```python
new_df['Customer_rating'].fillna(new_df['Customer_rating'].mean(), inplace = True)
new_df.isna().sum()
```

#### Data Cleaning pada atribut 'Product Importance'

##### Menggunakan modus / mode karena atribut Product Importance bertipe data object atau tulisan

```python
new_df['Product_importance'].fillna(new_df['Product_importance'].mode, inplace = True)
new_df.isna().sum()
```
