### Import Package

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import Dataset

```python
df = pd.read_csv('Train.csv')
df
```

### GRAFIK BATANG

```python
dfTrain = df['Mode_of_Shipment'].value_counts()

plt.figure(figsize=(15,10)) 

plt.bar(dfTrain.index, dfTrain, color = ['lavenderblush', 'lavender', 'lightcyan'])
plt.title ('Data Pengiriman Paket', fontweight = 'bold')
plt.xlabel ('Jalur Pengiriman', fontweight = 'bold')
plt.ylabel ('Jumlah', fontweight = 'bold')
plt.show()
```

#### Dari visualisasi data di atas dapat kita simpulkan bahwa pengiriman paket via jalur laut lebih banyak dibandingkan jalur udara dan darat.

### GRAFIK HISTOGRAM

```python
df['Cost_of_the_Product'].value_counts()
```

```python
plt.figure(figsize=(15,10))

plt.hist(df['Cost_of_the_Product'], color = 'thistle', bins = 20, rwidth=0.9)

plt.title ('Cost of the Product', fontweight = 'bold')
plt.xlabel ('Cost', fontweight = 'bold')
plt.ylabel ('Jumlah', fontweight = 'bold')
plt.show()
```

#### Data di atas adalah visualisasi data dari 'Cost of the Product' yang mana artinya adalah harga dari sebuah produk, kita memiliki data tertinggi dan terendah. Data tertinggi berada di rentang harga perkiraan 255 sampai harga 260. Yang artinya barang yang memiliki harga mencapai 255 - 260 lebih banyak jumlahnya dibanding data yang lain. Sedangkan data terendah berada pada rentang harga 220 - 230.

### PIE CHART

```python
dfRate = df['Customer_rating'].value_counts()
dfRate
```

```python
plt.figure(figsize=(8,8)) 
plt.pie(dfRate, 
        labels = dfRate.index,
        autopct = "%1.2f%%",
        startangle = 90, 
        explode = [0.1,0, 0, 0, 0],
        shadow = True) 
plt.title('Customer Ratings', fontweight = 'bold')
plt.show()
```

#### Dilihat dari pie chart di atas perbandingan data customer rating tidak jauh berbeda, namun customer lebih banyak memberi rating 3 sebanyak 20,36% atau sebanyak 2239 customer yang memberi rating 3 pada produk.

### BOX PLOT

```python
plt.figure(figsize=(15,10))
sns.boxplot (
    df,
    y = "Discount_offered",
    x = "Product_importance",
)

plt.title('Data antara seberapa pentingnya produk dan diskon yang ditawarkan', fontweight = 'bold')
plt.show()
```

#### Dari data di atas bisa kita peroleh bahwa low dan mediumnya produk memiliki rata rata yang sama yaitu berada di angka 5, bisa dilihat dari garis tengahnya dan juga Q1 dan Q3 pada low dan medium memiliki nilai yang sama. Sedangkan pada data high memiliki Q1 yang lebih kecil daripada Q3 dan juga memiliki nilai rata-rata yang berada di angka 5. Data ketiganya pun memiliki outliers yang banyak yang mana nantinya mungkin akan mempengaruhi nilai rata-rata maupun standar deviasi dari data jika outliers tersebut di analisis lebih lanjut.

### HEATMAP

```python
plt.figure(figsize=(15,10))

sns.heatmap (
    df.corr(),
    annot = True, #untuk memperlihatkan angkanya
    cmap = 'PRGn', #colormap (pallete warna)
)
plt.title ('Mencari Korelasi Data dengan Heatmap')
plt.show()
```

#### Dari visualisasi data di atas kita peroleh data dengan korelasi tertinggi yaitu pada 'Discount Offered' dan 'Reached on Time' dengan nilai 0.4, maka jika nilai pada 'Discount Offered' naik maka nilai pada 'Reached on Time' juga ikut naik karena mereka memiliki korelasi yang tinggi.
